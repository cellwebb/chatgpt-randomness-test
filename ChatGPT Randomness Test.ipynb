# Chat Completion - *gpt-3.5-turbo*

```python
model = "gpt-3.5-turbo"
messages = [{"role": "user", "content": "I'm thinking of a number between 1 and 10. The number is: "}]
temperature = 1
n = 10

chat_completion = openai.ChatCompletion.create(model=model, messages=messages, temperature=temperature, n=n)


responses = [x.message.content for x in chat_completion.choices]
print(responses)
```

```python
model = "gpt-3.5-turbo"
messages = [{"role": "user", "content": "Generate a random number between 1 and 10."}]
temperature = 1
n = 128

chat_completion = openai.ChatCompletion.create(model=model, messages=messages, temperature=temperature, n=n)

responses = [x.message.content for x in chat_completion.choices]
print(responses)
responses = [x[:-1] if x.endswith('.') else x for x in responses]  # remove trailing periods
print(responses)
sns.displot(sorted(responses))
```

```python
animals = ["dog", "cat", "mouse", "spider", "bird", "elephant", "giraffe", "kangaroo", "firefly", "moose"]

model = "gpt-3.5-turbo"
temperature = 1
n = 128
max_tokens = 1

for _ in range(4):
    random.shuffle(animals)
    print("Reshuffled list: " + ', '.join(animals))

    messages = [{"role": "user", "content": "Pick a random animal from: " + ', '.join(animals)}]
    temperature = 1
    n = 128
    max_tokens = 3

    chat_completion = openai.ChatCompletion.create(
        model=model, messages=messages, temperature=temperature, n=n, max_tokens=max_tokens)
    responses = [x.message.content.strip().lower() for x in chat_completion.choices]
    responses = [x[:-1] if x.endswith('.') else x for x in responses]  # remove trailing periods
    # print(responses)

    chart = sns.displot(sorted(responses))
    chart.set_xticklabels(rotation=45, horizontalalignment='right')
```
