```python
import os
import random
import openai
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
model = "text-davinci-003"
prompt = "I'm thinking of a number between one and ten. The number is: "

completion = openai.Completion.create(model=model, prompt=prompt, n=128)
responses = [x.text.strip() for x in completion.choices]

print(responses)
sns.displot(sorted(responses))
```
